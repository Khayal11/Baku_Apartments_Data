```python
#Remove duplicates from data
baku_unique_data = Baku_apart_data.drop_duplicates(ignore_index=True)
baku_unique_data
```

```python
# Split column Floor and create a new column for analysis 

baku_unique_data[['Current_floor','Total_floors']]=baku_unique_data['Floor'].astype(str).str.split('/',expand=True).astype(float)

baku_unique_data
```

```python
# Filter rows where old buildings and full floor is greater 9

check_old_build = baku_unique_data[(baku_unique_data['New_building']== 0 ) & (baku_unique_data['Total_floors']>9)].index

check_old_build
```

```python
# Remove filtered rows

baku_unique_data.drop(check_old_build,inplace=True)

baku_unique_data
```

```python
# Create new column by exctracting the last letters and adding it
baku_unique_data['Zone'] = baku_unique_data['Location'].str.split().str[-1]
baku_unique_data
```

```python
# Replace abbreviations with real words
baku_unique_data['Zone'] =  baku_unique_data['Zone'].replace({'m.' : 'Metro', 'r.' : 'Region' , 'q.' : 'Settlement'})

baku_unique_data
```

```python
# Top 10 apartment's location with the highest price
top_price = baku_unique_data.groupby(['Location','Zone'])['Price'].max().sort_values(ascending=False).reset_index()
top_10_price = top_price.head(10)
top_10_price
```

```python
# Plotting top 10 expensive locations with prices
top_10_price.plot(kind = 'bar', x= 'Location', y = 'Price',legend=False,figsize=(10,8))
plt.title('Top 10 Locations with Highest Price')
plt.xlabel('Location')
plt.ylabel('Price (mln)')
plt.show()
```

```python
# Add new column to get price per square meter for location
baku_unique_data['Price_per_square'] = baku_unique_data['Price'] / baku_unique_data['Square']

baku_unique_data.head(10)
```

```python
# Find the average price per suqare meter by location

avg_price_per_square = baku_unique_data.groupby('Location')['Price_per_square'].mean().sort_values(ascending=False)

avg_price_per_sq_top = avg_price_per_square.head(10)

avg_price_per_sq_top
```

```python
# Plot top 10 locations with highest price per square meter
avg_price_per_sq_top=avg_price_per_sq_top.sort_values(ascending=False)

ax = avg_price_per_sq_top.plot(kind='barh', figsize=(10,6),color='orange')


plt.title('Top 10 Locations with the Highest Price per Square Meter')
plt.xlabel('Price per Square Meter (AZN)')
plt.ylabel('Location')
plt.grid(axis='x', linewidth = 1,color = 'green')

for i in ax.patches:
    ax.text(i.get_width() + 100, i.get_y() + 0.3, f"{i.get_width():.2f}", fontsize = 10 ,color = 'Black')


plt.show()


```

```python
# Filter for 1 room, 2 rooms ,3 rooms and 4 rooms apartments in old buildings

one_room_old_build = baku_unique_data.loc[(baku_unique_data['Rooms'] == 1) & (baku_unique_data['New_building'] == 0)]

two_room_old_build = baku_unique_data.loc[(baku_unique_data['Rooms'] == 2) & (baku_unique_data['New_building'] == 0)]

three_room_old_build = baku_unique_data.loc[(baku_unique_data['Rooms'] == 3) & (baku_unique_data['New_building'] == 0)]

four_room_old_build = baku_unique_data.loc[(baku_unique_data['Rooms'] == 4) & (baku_unique_data['New_building'] == 0)]

```

```python
# Find the highest price rooms in old buildings

high_one_old_build = one_room_old_build.loc[one_room_old_build['Price'].idxmax()]
loc_1_old_high= high_one_old_build['Location']
price_1_old_high =high_one_old_build['Price']
print(f"The Location with the highest price for 1 room apartment in old buildings is near {loc_1_old_high}, with a price of {price_1_old_high} AZN.")


high_two_old_build = two_room_old_build.loc[two_room_old_build['Price'].idxmax()]
loc_2_old_high= high_two_old_build['Location']
price_2_old_high =high_two_old_build['Price']
print(f"The Location with the highest price for 2 rooms apartment in old buildings is near {loc_2_old_high}, with a price of {price_2_old_high} AZN.")

high_three_old_build = three_room_old_build.loc[three_room_old_build['Price'].idxmax()]
loc_3_old_high= high_three_old_build['Location']
price_3_old_high =high_three_old_build['Price']
print(f"The Location with the highest price for 3 rooms apartment in old buildings is near {loc_3_old_high}, with a price of {price_3_old_high} AZN.")

high_four_old_build = four_room_old_build.loc[four_room_old_build['Price'].idxmax()]
loc_4_old_high= high_four_old_build['Location']
price_4_old_high =high_four_old_build['Price']
print(f"The Location with the highest price for 4 rooms apartment in old buildings is near {loc_4_old_high}, with a price of {price_4_old_high} AZN.")


```

```python
# Show the highest prices for rooms both in old and new buildings

rooms_total = ['1 room', '2 rooms', '3 rooms', '4 rooms']

new_building_prices = [price_1_high , price_2_high, price_3_high, price_4_high]
old_building_prices = [price_1_old_high, price_2_old_high, price_3_old_high, price_4_old_high]

new_building_locations = [loc_1_high, loc_2_high, loc_3_high, loc_4_high]
old_building_locations = [loc_1_old_high, loc_2_old_high, loc_3_old_high, loc_4_old_high]

# Bar chart configuration

x = np.arange(len(rooms_total))
width = 0.35

fig, ax = plt.subplots(figsize=(10,6))

a1 = ax.bar(x - width/2, new_building_prices ,width , label = 'New_building')
a2 = ax.bar(x + width/2, old_building_prices ,width , label = 'Old_building')

# Set labels

ax.set_xlabel('Number of Rooms')
ax.set_ylabel('Price(Azn)')
ax.set_title('Price Comparison for New and Old Buildings with Locations')
ax.set_xticks(x)
ax.set_xticklabels(rooms_total)
ax.legend()

# Add location labels above each bar
for i, a in enumerate(a1):
    ax.text(a.get_x() + a.get_width()/2., a.get_height(),new_building_locations[i],ha='center',va='bottom')

for i, a in enumerate(a2):
    ax.text(a.get_x() + a.get_width()/2., a.get_height(),old_building_locations[i],ha='center',va='bottom')

plt.tight_layout()
plt.grid()
plt.show()

 

```

```python
# Average,min,max price by floor

price_analysis_by_floor = baku_unique_data.groupby('Current_floor')['Price'].agg(['max','min','mean']).reset_index()

price_analysis_by_floor
```

```python
# Show the relationships between floor levels and prices

plt.figure(figsize=(10,6))

plt.plot(price_analysis_by_floor['Current_floor'],price_analysis_by_floor['mean'],label = 'Average Price', marker='o',color='blue')
plt.plot(price_analysis_by_floor['Current_floor'],price_analysis_by_floor['min'], label = 'Min Price', marker='o',linestyle = '--',color ='green')
plt.plot(price_analysis_by_floor['Current_floor'],price_analysis_by_floor['max'], label = 'Max Price', marker = 'o',linestyle = '--',color = 'red')

plt.title('Price Trends by Floor')
plt.xlabel('Floor Level')
plt.ylabel('Price (mln)')
plt.grid(True)
plt.show()
```

```python
# Price Comparision for properties with/without Bill of Sale and Mortgage to determine if mentioned properties tend to be more expensive 

fig, ax = plt.subplots(1,2,figsize = (12,6))

sns.boxplot(data=baku_unique_data, x= 'Has_bill_of_sale',y= 'Price',ax = ax[0],color='red',legend=False)
ax[0].set_title('Price Comparision: Bill of Sale')
ax[0].set_xlabel('Has Bill od Sale')
ax[0].set_ylabel('Price')

sns.boxplot(data= baku_unique_data, x='Has_mortgage', y= 'Price', ax= ax[1],color='green', legend=False)
ax[1].set_title('Price Comparision: Mortgage')
ax[1].set_xlabel('Has Mortgage')
ax[1].set_ylabel('Price')

plt.tight_layout()
plt.show()
```
